```python
# Example: reuse your existing OpenAI setup
from openai import OpenAI
import json
import orjson
```

```python
with open('german_rag_transformed_reformated_MS-MARCO_v1.1.jsonl', 'r', encoding='utf-8', errors='ignore') as file:
    lines = [orjson.loads(line) for line in file.readlines()[:10] if line.strip()]
```

```python
# Point to the local server
client = OpenAI(base_url="http://localhost:1234/v1", api_key="lm-studio")

system_prompt = """
Du bist ein Programmierassistent und unterstützt bei der Erstellung von Trainingsdaten für ein Frage-Antwort-Modell.

Du erhältst ein JSON-Objekt (ein Python-Dictionary) mit den Feldern `query`, `context`, ... . Das Feld `context` enthält eine Liste von Textausschnitten (Strings), die potenziell relevante Informationen zur Beantwortung der `query` enthalten.

Deine Aufgabe ist es, eine präzise und vollständige Antwort im Feld `response` zu generieren. Verwende ausschließlich Informationen aus dem gegebenen `context`. Formuliere die Antwort in deinen eigenen Worten auf Grundlage der relevanten Sätze. Am Ende der Antwort gibst du die verwendeten Sätze als Indexnummern in eckigen Klammern an (z. B. `[2][3]`), entsprechend ihrer Position in der `context`-Liste. Gib zusätzlich die Liste dieser ausgewählten Indizes im Feld `selected_sents` an.

Das Ausgabeformat soll **genau** wie im folgenden Beispiel aussehen:

<Beispiel>
{
  "query": "Welche Bedeutung hatten The Melvins für die Entstehung des Grunge-Genres in Seattle in den 1980er Jahren?",
  "context": [
    "Mitte der 1980er bildete sich ein Kern in der Szene Seattles, zu dem neben Green River, Soundgarden und The U-Men auch The Melvins gehörten",
      "Zu diesem Zeitpunkt entstand auch die Bezeichnung \"Grunge\" als Genrebegriff.",
      "Die Kommerzialisierung der Szene ging schnell vonstatten",
      "Markante Bekleidungsstücke wie das Flanellhemd wurden als neue Mode verkauft",
      "So wurde ein banaler Alltagsgegenstand zum Dresscode der Grunge-Szene und drang in den Mainstream vor."
  ],
  "selected_sents": [
    0
  ],
  "response": "The Melvins gehörten zu einem Kern der Szene Seattles, die maßgeblich an der Entstehung des Grunge-Genres beteiligt war. [0]"
}
</Beispiel>

**Wichtige Regeln:**
- Nutze ausschließlich Informationen aus dem `context`, keine externen Quellen oder eigenes Vorwissen.
- Verwende eine sachliche, neutrale Sprache.
- Gib **nur** das JSON-Objekt zurück – ohne zusätzliche Kommentare, Erklärungen oder Text.

Das Ergebnis wird zur Erstellung einer `.jsonl`-Datei für das Training eines Sprachmodells verwendet.
"""

structured_output_data = {
  "type": "json_schema",
  "json_schema": {
    "name": "formatted_data",
    "schema": {
      "type": "object",
      "properties": {
        "query": {
          "type": "string",
          "description": "`query` from the input dictionary."
        },
        "context": {
          "type": "array",
          "description": "`passages.passage_text` from the input dictionary.",
          "items": {
            "type": "string"
          }
        },
        "selected_sents": {
          "type": "array",
          "description": "The list of indexes corresponding to the contexts from `context` you will use to generate the response to `query`.",
          "items": {
            "type": "integer",
            "minimum": 1
          }
        },
        "response": {
          "type": "string",
          "description": "The answer to `query` generated using `context`. The sources (indexes from `selected_sents`) should be included at the end."
        }
      },
      "required": ["query", "context", "selected_sents", "response"]
    }
  }
}

  
formatted_jsons = []

for line in lines:
  messages=[
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": f"{json.dumps(line, ensure_ascii=False)}"}
  ]

  completion = client.chat.completions.create(
    model="model-identifier",
    messages=messages,
    temperature=0.7,
    response_format=structured_output_data,
    stream=False
  )

  formatted_jsons.append(json.loads(completion.choices[0].message.content, strict=False))
```

```python
print(json.dumps(formatted_jsons, indent=2, ensure_ascii=False))

# with open('formatted_output.json', 'w', encoding='utf-8') as f:
#     f.write(json.dumps(formatted_jsons, ensure_ascii=False, indent=2))
```
